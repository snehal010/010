```python
# removing the unnecessary columns
df = df.drop(columns=['Id'])
```

```python
#feature and target seperation
x = df.drop(columns=['Price'])
y = df['Price']

#x conatains all input feature
#y contains the target variable or what we want to predict
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(
    x, y, test_size=0.2,random_state=42
)
```

```python
#linear models and requiarlization are sensitive featureypes
#Scalar to before regrassion we encode categoricalls
#and scale numericals
```

```python
categorical_cols = ['Location','Condition','Garage']
numerical_cols = x_train.drop(columns=categorical_cols).columns
```

```python
categorical_cols, numerical_cols
```

```python
#we will use oneHotEncoder cuz regrassion needs numbers
#it increases features set up regularization need

from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer

#ColumTransformer is to apply diffrent preprosessing steps to
#diffrent columns in one go because
#numerical coals need scaling , categorical_cols need encoding
#so we should not treat all columns the same

```

```python
preprocessor = ColumnTransformer(transformers=[
    ('num', StandardScaler(),numerical_cols),
     ('cat', OneHotEncoder(drop='first'),categorical_cols)])

#transformers =[]
#this is list of rules
#each rule follow this feather_format(name)
```

```python
x_train_processed = preprocessor.fit_transform(x_train)
x_test_processed = preprocessor.transform(x_test)
```

```python
x_train_processed.shape
```

```python
x_test_processed.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train_processed,y_train)
```

```python
y_train_pred = lr.predict(x_train_processed)
y_test_pred = lr.predict(x_test_processed)

#predicting on both so
#train prediction - check how well model learned
#test prediction - check how well model generalizes

#we predict on training data to see learning in test to
#see the real world performance

#y_train_pred, y_test_pred
```

```python
from numpy import square
#error and residual interpretation

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

# the reason for using,mae,mse,rmse,r2 is that we never judge a
# model by accuracy in LinearRegression
# herewe used error based metrics

mae=mean_absolute_error(y_test,y_test_pred)
mse=mean_squared_error(y_test,y_test_pred)
rmse=np.sqrt(mse)
r2=r2_score(y_test,y_test_pred)

print("MAE",mae)
print("MSE",mse)
print("RMSE",rmse)
print("R2",r2)

# MAE tellls the avg absolute difffrence between predicted
# and actual house prices
# we use this because its easy to understand , same unit as price ,
# less sensitive to outliers(an outlier is a value that lies far away
# from the majority of values in the same cal)

# MSE - its squares the errors, so large mistakes are penalized more
# it highlights big prediction erroes, used by many algorithms

# RMSE - it brings the error back to the original price unit , make_column_transformer
# it easier to cannect or interpret, penalizes large erroes, same unit
# comparing target_outdated

# r2-R2 tells us how much of the variation in house prices is explained
# by the model
```

```python
#r2
#0.0 - model explins nothing
#1.0 - perfectimodel
#0.80 - model explin 80% of the variation
#r2 < 0 - worse than avg prediction
```

```python
train_r2 = r2_score(y_train,y_train_pred)
test_r2 = r2_score(y_test,y_test_pred)

print("Train R2:",train_r2)
print("Test R2:",test_r2)

# eventhough our error values look reasonable , the r2 score is
# negative , wich tells us the model is not learning meaningful pattern
#this indicate underfitting and we need regularization or a better model
```

```python
# Regularization (ridge and lasso)
# ridge regression

from sklearn.linear_model import Ridge
ridge=Ridge(alpha = 1.0)
ridge.fit(x_train_processed,y_train)

y_train_pred=ridge.predict(x_train_processed)
y_test_pred=ridge.predict(x_test_processed)

#ridge reduces the ipmpact of useless features
```

```python
from sklearn.metrics import r2_score

ridge_train_r2 = r2_score(y_train, y_train_pred)
ridge_test_r2 = r2_score(y_test, y_test_pred)

ridge_train_r2, ridge_test_r2
```

```python
#lasso regrassion
#ridge reduces the impact of useless feature while lasso can completly
#eliminate unless features

from sklearn.linear_model import Lasso

lasso = Lasso(alpha=1.0)
lasso.fit(x_train_processed,y_train)

y_train_pred_lasso = lasso.predict(x_train_processed)
y_test_pred_lasso = lasso.predict(x_test_processed)

#ridge - shrinks coefficient
#lasso - shrinks and can remove coefficients sometimes make them zero too
```

```python
lasso_train_r2=r2_score(y_train,y_train_pred_lasso)
lasso_test_r2=r2_score(y_test,y_test_pred_lasso)

lasso_train_r2,lasso_test_r2
```

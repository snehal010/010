```python
from numpy import histogram
# matplotib
# matplotib tool is helpful for visualization
# histogram
import matplotlib.pyplot as plt
plt.figure()
plt.hist(df['G3'], bins = 10)
plt.xlabel("G3")
plt.ylabel("Frequency")
plt.title("Histogram of G3")
plt.show()
```

```python
df.columns
```

```python
# barchat
avg_score_by_gender = df.groupby('sex')['G3'].mean()
avg_score_by_gender

plt.figure()
avg_score_by_gender.plot(kind='bar')
plt.xlabel("Gender")
plt.ylabel("Average Final Mark")
plt.title("Avrage Final Mark by Gender")
plt.show()
```

```python
# scatter plot
plt.figure()
plt.scatter(df['studytime'],df['G3'])
plt.xlabel('study time')
plt.ylabel('G3')
plt.title('scatter plot of G1 vs G3')
plt.show()
```

```python
#box plot
plt.figure()
plt.boxplot(df['G3'])
plt.ylabel('G3')
plt.title('Box Plot of G3')
plt.show()
```

```python
# box plot
plt.figure()
plt.boxplot(df['G3'])
plt.ylabel('G3')
plt.title("spread of final scores")
plt.show()
```

```python
#line plots

avg_marks = df[['G1','G2','G3']].mean()

plt.figure()
plt.plot(avg_marks.index, avg_marks.values, marker = 'o')
plt.ylabel("Average marks")
plt.title("Average marks of Students")
plt.show()
```

```python
#with bar chart we will find the number of students
#who failed and passed the final exam

passed = (df['G3']>= 10).sum()
failed = len(df) - passed
plt.figure()
plt.bar(['Pass', 'Fail'], [passed, failed])
plt.ylabel("Number of students")
plt.title("Number of students who passed and failed")
plt.show()
```

```python
plt.figure()
plt.hist(df['absences'],bins=10,color='skyblue',edgecolor='red')
plt.xlabel("Absences")
plt.ylabel("Frequency")
plt.title("Histogram of Absences")
plt.show()
```
